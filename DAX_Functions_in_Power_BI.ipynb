#**DAX Functions in Power BI**

                                  
1.Calculate column Totalprice using Quantity and unit Price.

Ans:  TotalPrice = 'Table'[Quantity] * 'Table'[UnitPrice]

2.Create a calculated column DiscountedPrice.

Ans: DiscountedPrice =
'Table'[TotalPrice] - ('Table'[TotalPrice] * 'Table'[Discount] / 100)

3.Create a measure Average Order Value (AOV).

Ans: Total Sales = SUM('Table'[SalesAmount]).
YTD Sales = TOTALYTD( SUM('Table'[SalesAmount]),
  'Table'[OrderDate])

4.Using DAX, create a measure that calculates cumulative Year-to-Date (YTD) Sales based on the Order Date.

Ans: YTD Sales =
TOTALYTD(
    SUM('Table'[SalesAmount]),
    'Table'[OrderDate]
)

5.Using DAX, create a measure to rank each product according to its Total Sales, with the highest-selling product ranked first.

Ans: Product Rank =
RANKX(
    ALL('Table'[ProductID]),
    CALCULATE(SUM('Table'[SalesAmount])),
    ,
    DESC,
    DENSE
)

6.Create a calculated column classifying orders as “High Value” if sales > 2000, else “Low Value”.

Ans: Order Value Category =
IF(
    'Table'[SalesAmount] > 2000,
    "High Value",
    "Low Value"
)

7.Create a measure to calculate total sales from the Online channel only.

Ans: Online Sales =
CALCULATE(
    SUM('Table'[SalesAmount]),
    'Table'[SalesChannel] = "Online"
)

8.Ceate a measure called "Date Difference" that calculates the number of days between the latest  Order Date and today's date using the DATEDIFF function.

Ans: Date Difference =
DATEDIFF(
    MAX('Table'[OrderDate]),
    TODAY(),
    DAY
)
